# Semana 16: Planejando a Estrutura da API

Nesta semana, vamos planejar a estrutura da nossa API. Discutiremos as melhores práticas para organizar os endpoints, definir os recursos e planejar as interações entre cliente e servidor.

--- 
**⚠️ ATENÇÃO: AULA PRÁTICA NO SEU COMPUTADOR!**

Continue trabalhando no seu projeto Django local, com o ambiente virtual `(venv)` ativado.

--- 
### 1. Identificando os Requisitos da API

Antes de começar a implementar a API, é importante entender os requisitos funcionais e não funcionais do projeto. Faça as seguintes perguntas:

- Quais recursos a API deve expor?
- Quais operações (CRUD) são necessárias para cada recurso?
- Quais dados devem ser retornados em cada endpoint?
- Quais restrições de segurança devem ser aplicadas?

### Passo 1: Planejando os Endpoints

Com base nos requisitos, planeje os endpoints necessários. Por exemplo:

| Recurso   | Método HTTP | Endpoint         | Descrição                     |
|-----------|-------------|------------------|--------------------------------|
| Produtos  | GET         | `/api/produtos/` | Lista todos os produtos       |
| Produtos  | POST        | `/api/produtos/` | Cria um novo produto          |
| Produtos  | GET         | `/api/produtos/<id>/` | Detalha um produto específico |
| Produtos  | PUT         | `/api/produtos/<id>/` | Atualiza um produto específico|
| Produtos  | DELETE      | `/api/produtos/<id>/` | Remove um produto específico  |

### Passo 2: Criando o Documento de Planejamento

Com os endpoints definidos, crie um documento detalhado que inclua:

1. **Descrição dos Recursos:**
    - Produtos: Representa os itens disponíveis na loja.

2. **Estrutura de Dados:**
    ```json
    {
        "id": 1,
        "nome": "Smartphone",
        "preco": 1999.99,
        "categoria": "Eletrônicos"
    }
    ```

3. **Fluxo de Interação:**
    - O cliente faz uma requisição GET para `/api/produtos/` para listar os produtos.
    - O cliente faz uma requisição POST para `/api/produtos/` para adicionar um novo produto.

### Conclusão

Com o planejamento detalhado, você terá um guia claro para implementar a API. Nas próximas aulas, vamos começar a codificar os endpoints e testar as interações planejadas.
